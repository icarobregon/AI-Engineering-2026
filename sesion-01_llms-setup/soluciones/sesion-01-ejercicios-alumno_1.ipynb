## Nivel 1 — Llamada básica (obligatorio)

Envía un mensaje al modelo y muestra la respuesta por pantalla.

**Elige la sección de tu proveedor y ejecuta el código.**

### Opción A: OpenAI

```python
response = openai_client.responses.create(
    model="gpt-4o-mini",
    input="Explain in two sentences what a REST API is."
)

print(response.output_text)
```

### Opción B: Anthropic

```python
response = anthropic_client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": "Explain in two sentences what a REST API is."}
    ]
)

print(response.content[0].text)
```

## Nivel 3 — Metadatos y coste (opcional)

Extrae de la respuesta del modelo:
- Tokens de entrada y salida
- Modelo utilizado
- Coste estimado de la llamada

Consulta las tarifas de tu proveedor:
- OpenAI: https://openai.com/api/pricing
- Anthropic: https://docs.anthropic.com/en/docs/about-claude/models

### Opción A: OpenAI

```python
response = openai_client.responses.create(
    model="gpt-4o-mini",
    instructions="You are a software estimation expert.",
    input="Explain in two sentences what a REST API is."
)

print("=== RESPONSE ===")
print(response.output_text)

print("\n=== METADATA ===")
print(f"Model: {response.model}")
print(f"Status: {response.status}")
print(f"Input tokens: {response.usage.input_tokens}")
print(f"Output tokens: {response.usage.output_tokens}")
print(f"Total tokens: {response.usage.total_tokens}")

# gpt-4o-mini pricing (check for updated rates)
INPUT_PRICE_PER_MILLION = 0.15
OUTPUT_PRICE_PER_MILLION = 0.60

input_cost = (response.usage.input_tokens / 1_000_000) * INPUT_PRICE_PER_MILLION
output_cost = (response.usage.output_tokens / 1_000_000) * OUTPUT_PRICE_PER_MILLION
total_cost = input_cost + output_cost

print(f"\n=== ESTIMATED COST ===")
print(f"Input cost:  ${input_cost:.6f}")
print(f"Output cost: ${output_cost:.6f}")
print(f"Total cost:  ${total_cost:.6f}")
```

### Opción B: Anthropic

```python
response = anthropic_client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=1024,
    system="You are a software estimation expert.",
    messages=[
        {"role": "user", "content": "Explain in two sentences what a REST API is."}
    ]
)

print("=== RESPONSE ===")
print(response.content[0].text)

print("\n=== METADATA ===")
print(f"Model: {response.model}")
print(f"Stop reason: {response.stop_reason}")
print(f"Input tokens: {response.usage.input_tokens}")
print(f"Output tokens: {response.usage.output_tokens}")

# claude-haiku-4-5 pricing (check for updated rates)
INPUT_PRICE_PER_MILLION = 1.00
OUTPUT_PRICE_PER_MILLION = 5.00

input_cost = (response.usage.input_tokens / 1_000_000) * INPUT_PRICE_PER_MILLION
output_cost = (response.usage.output_tokens / 1_000_000) * OUTPUT_PRICE_PER_MILLION
total_cost = input_cost + output_cost

print(f"\n=== ESTIMATED COST ===")
print(f"Input cost:  ${input_cost:.6f}")
print(f"Output cost: ${output_cost:.6f}")
print(f"Total cost:  ${total_cost:.6f}")
```
